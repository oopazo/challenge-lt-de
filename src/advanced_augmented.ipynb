```python
# defining index collection name:
collection = "advanced_augmented_index"
if not collection_exists(collection):
    # reading files from the directory
    documents = read_files()
    print(f"{len(documents)} documents.")
    # parse documents, generates multiple nodes:
    nodes = parse_documents(documents)
    print(f"Chunked into {len(nodes)} nodes.")
    # improving nodes structure
    nodes = organize_node_structure(nodes)
    # augment nodes with metadata using a simple approach:
    nodes = augment_metadata(nodes)
    # chunking nodes into smaller pieces and adding metadata augmentation using llms:
    nodes = chunk_advanced_augmented_metadata(nodes)
    # creating vector store index
    index = get_or_create_chroma_index(nodes=nodes, collection=collection, mode="create")
else:
    index = get_or_create_chroma_index(nodes=[], collection=collection, mode="load")
```
